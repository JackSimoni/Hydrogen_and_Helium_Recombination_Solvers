```python
# Ryden 8.1

import numpy as np
import matplotlib.pyplot as plt

E_ion = 13.6 * 1.6e-19  # ionization energy of H [J]
k_b = 1.38e-23 # J/K
m_e = 9.11e-31 # kg
c_light = 3e8 # m/s

def saha_equation(T, eta):
    
    S = 3.84 * eta * ((k_b*T)/(m_e*c_light**2))**1.5 * np.exp(E_ion/(k_b*T))
    X = (-1 + np.sqrt(1 + 4*S)) / (2*S)
    
    return X


T_vals = np.linspace(3000, 4500, 1000000)
etas = [4e-10, 8e-10]
X_vals = {eta: [saha_equation(T, eta) for T in T_vals] for eta in etas}

# Plotting 
plt.figure(figsize=(8, 6))
for eta in etas:
    plt.plot(T_vals, X_vals[eta], label=f"$\eta = {eta:.0e}$")

plt.xlabel("Temperature [K]")
plt.ylabel("Ionization Fraction $X$")
plt.title("Hydrogen Recombination Fraction vs Temperature")
plt.legend()
plt.grid(True)
plt.show()

def find_Trec(T_vals, X_vals, threshold=0.5):
    idx = np.argmin(np.abs(np.array(X_vals) - threshold))
    return T_vals[idx]

for eta in etas:
    Trec = find_Trec(T_vals, X_vals[eta])
    print(f"T_rec for Î· = {eta:.0e} is approximately {Trec:.0f} K")
```

As can be seen, changing the photon-baryon ratio by a factor of 2, does not have a significant effect on the recombination tempertaure.

```python
# Ryden 8.2

import numpy as np
import matplotlib.pyplot as plt

# Constants
E_ion = 13.6 * 1.6e-19  # ionization energy of H [J]
k_b = 1.38e-23 # Boltzmann constant [J/K]
m_e = 9.11e-31 # electron mass [kg]
c_light = 3e8  # speed of light [m/s]

# Function: number of photons above threshold energy
def photon_number_threshold(T, threshold=E_ion):
    normalization = (6.1e-10)
    y = 0.42 / normalization * (threshold / (k_b * T))**2 * np.exp(-threshold / (k_b * T))
    return y

T_vals = np.linspace(4000, 6000, 1000000)

Y_vals = [photon_number_threshold(T) for T in T_vals]

# Plot
plt.figure(figsize=(8, 6))
plt.plot(T_vals, Y_vals, label="Photon number threshold curve")

plt.xlabel("Temperature [K]")
plt.ylabel("Photon Number above threshold")
plt.title("Photon Number Above Ionization Energy vs Temperature")
plt.legend()
plt.grid(True)
plt.show()

def find_Trec(T_vals, Y_vals, threshold=1.0):
    idx = np.argmin(np.abs(np.array(Y_vals) - threshold))
    return T_vals[idx]

Trec = find_Trec(T_vals, Y_vals, threshold=1.0)
print(f"T_rec (for Photon Number Above 13.6 eV Threshold = 1.0) â {Trec:.0f} K")
```
